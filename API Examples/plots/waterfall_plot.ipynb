```python
import xgboost
import shap
```

```python
# train XGBoost model
X, y = shap.datasets.adult()
model = xgboost.XGBClassifier().fit(X, y)
```

```python
# compute SHAP values
explainer = shap.Explainer(model, X)
shap_values = explainer(X)
```

瀑布圖是用來顯示單一預測的解釋

```python
# 單筆資料的解釋
shap.plots.waterfall(shap_values[0])
```

使用 max_display 參數更改 10 行的預設限制

```python
shap.plots.waterfall(shap_values[0], max_display=20)
```

擁有 2,174 美元的資本利得會大大降低此人每年賺取超過 5 萬美元的預測機率。由於瀑布圖僅顯示單一樣本數據，因此我們無法看到資本收益變化的影響。

我們可以使用散點圖，它顯示了資本利得的低值比沒有資本利得對收入的預測更為負面。

```python
shap.plots.scatter(shap_values[:, "Capital Gain"])
```
